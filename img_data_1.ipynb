```python
NNH = KNeighborsClassifier(n_neighbors= 5 , weights = 'distance' )
```

```python
bc_df = pd.read_excel("Dataset.xlsx")
```

```python
bc_df1=bc_df.drop(labels= "ImgID", axis=1)
```

```python
bc_df2=bc_df1.drop(labels= "Matching Percentage" , axis=1)
```

```python
img_data=bc_df2.transpose()
```

```python
img_data1=img_data.drop(labels= 0 , axis = 1)
img=img_data1.transpose()
img
```

```python
img_1=img.replace({"Matching":"Y"},1,regex=True)

img_2=img_1.replace({"Matching":"N"},0,regex=True)
img_2
```

```python
img_2.dtypes
```

```python
encoder=ce.OneHotEncoder(cols=['BodySize','Complexion','Occasion','DressColor','DressSize','DressDesign','DressCategory','WaistSize','Gender'],handle_unknown='return_nan',return_df=True,use_cat_names=True)
```

```python
data_encoded = encoder.fit_transform(img_2)
data_encoded
```

```python
data_encoded.describe()
```

```python
data_encoded.groupby(["Matching"]).count()
```

```python
bc_feature_df = data_encoded.drop(labels= "Matching" , axis = 1)
bc_feature_df.head()
bc_feature_df.dtypes
```

```python
bc_feature_df_z = bc_feature_df.apply(zscore)  # convert all attributes to Z scale 

bc_feature_df_z.describe()
```

```python
X = np.array(bc_feature_df_z)
X.shape
```

```python
bc_labels = img_2["Matching"]
```

```python
y = np.array(bc_labels)
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
NNH.fit(X_train, y_train)
```

```python
predicted_labels = NNH.predict(X_test)
NNH.score(X_test, y_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, predicted_labels))
```
